```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_features=2, n_samples=500, centers=5, random_state=3)
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(X,y)
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker='*', color="orange")
plt.show()
```

```python
from sklearn.datasets import make_circles, make_moons
X,y = make_moons(n_samples=200, noise=0.10)
```

```python
plt.scatter(X[:,0], X[:,1],c=y)
```

```python
from sklearn.cluster import DBSCAN
dbs = DBSCAN(eps=0.2, min_samples=5)
```

```python
dbs.fit(X)
```

```python
Y = dbs.fit_predict(X)
```

```python
print(Y)
```
